```python
import numpy as np
import matplotlib.pyplot as plt
import random
from ipywidgets import interact

#%matplotlib ipympl              #comment this line if using in google colab or install ipympl from pip
```

# Code

```python
class Particle():
    def __init__(self, x, y, vx, vy):
        self.pos = np.array([x, y])
        self.local_best = np.array([x, y])
        self.velocity = np.array([vx, vy])
    
    def update_lb(self, f):
        if f(self.pos[0], self.pos[1]) < f(self.local_best[0], self.local_best[1]):
            self.local_best = self.pos
    
    def update_pos(self):
        self.pos = self.pos + self.velocity
    
    def update_velocity(self, w, c1, r1, c2, r2, gb):
        self.velocity = w*self.velocity + c1*r1*(self.local_best-self.pos) + c2*r2*(gb-self.pos)
    
```

```python
class Swamp():
    def __init__(self, f, xbound, ybound, n):
        self.n = n
        self.particles = [Particle(random.uniform(xbound[0], xbound[1]), random.uniform(ybound[0], ybound[1]), 2*random.random(), 2*random.random()) for _ in range(n)]
        self.gb = self.particles[0].local_best
        self.f = f

    def update_gb(self):
        self.gb = self.particles[0].local_best
        for particle in self.particles:
            if self.f(*particle.local_best) < self.f(*self.gb):
                self.gb = particle.local_best
    
    def train(self, w, c1, r1, c2, r2, n_gen):
        history = []
        for _ in range(n_gen):
            temp_pos = []
            for particle in self.particles:
                particle.update_velocity(w, c1, r1, c2, r2, self.gb)
                particle.update_pos()
                particle.update_lb(self.f)
                self.update_gb()
                temp_pos.append(particle.pos)
            history.append(np.array(temp_pos))
        return np.array(history)
```

# Example

```python
def objective_function(x, y):
    return (x-0.5)**2 + (y-1)**2

xb = (-4, 4)
yb = (-4, 4)

feature_x = np.arange(xb[0], xb[1], 0.2)
feature_y = np.arange(yb[0], yb[1], 0.2)
[X, Y] = np.meshgrid(feature_x, feature_y)

plt.close()
plt.contour(X, Y, objective_function(X, Y))
plt.scatter([0.5], [1], marker="*")
```

```python
sw = Swamp(objective_function, xb, yb, 10)
r = sw.train(1, 0.8, 1, 0.8, 1, 10)                   #Change peer pressure from 0.3 to 0.1
```

```python
@interact
def show_gen(gen=(0, 9, 1)):
    plt.close()
    feature_x = np.arange(xb[0], xb[1], 0.2)
    feature_y = np.arange(yb[0], yb[1], 0.2)
    [X, Y] = np.meshgrid(feature_x, feature_y)
    plt.contour(X, Y, objective_function(X, Y))
    plt.scatter([0.5], [1], marker="*")
    plt.scatter(r[gen][:, 0], r[gen][:, 1])
    
```
